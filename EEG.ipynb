```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
```

```python
import eegtools.io

idle = eegtools.io.load_edf('Team B/Olya_idle.edf')
print idle
```

```python
RHYTHMS = ['AF3', 'F7', 'F3', 'FC5', 'T7', 'P7', 'O1', 'O2', 'P8', 'T8', 'FC6', 'F4', 'F8', 'AF4']

def plot_channels(data, channels):
    plt.figure(figsize=(20, 12))
    x = data.time
    for channel in channels:
        y = data.X[data.chan_lab.index(channel)]
        plt.plot(x, y, label=channel)
    plt.legend()
    plt.xlabel('time [s]')
    plt.ylabel('voltage')
    plt.show()
```

Let's plot an idle state signal:

```python
plot_channels(idle, RHYTHMS)
```

Then let's plot a signal obtained when the subject has closed eyes:

```python
closed_eyes = eegtools.io.load_edf('Team B/Olya_close_eyes.edf')
plot_channels(closed_eyes, RHYTHMS)
```

# 4

Let's compare the power spectral densities at several channels. Channel AF4:

```python
from scipy import signal

def subplot_single_psd(data, label, channel):
    f, Pxx_den = signal.welch(data.X[data.chan_lab.index(channel)], data.sample_rate)
    plt.plot(f[8:], Pxx_den[8:], label=label)

def plot_comparison_psd(data_first, label_first, data_second, label_second, channel):
    plt.figure(figsize=(20, 12))
    subplot_single_psd(data_first, label_first, channel)
    subplot_single_psd(data_second, label_second, channel)
    plt.xlabel('frequency [Hz]')
    plt.ylabel('PSD [V**2/Hz]')
    plt.legend()
    plt.plot()

def compare_open_closed_eyes(channel):
    plot_comparison_psd(idle, 'open eyes', closed_eyes, 'closed eyes', channel)    

compare_open_closed_eyes('AF4')
```

Let's check it on other channels:

```python
compare_open_closed_eyes('O1')
```

```python
compare_open_closed_eyes('F3')
```

```python
compare_open_closed_eyes('P7')
```

```python
compare_open_closed_eyes('T8')
```

Hooray! We see a clear rise on alfa-band (7.5-12.5 Hz) when the eyes are closed on channels O1, F3, P7 (left hemisphere of the brain). The two channels that don't have any significant difference in alfa-channel are both located on the right hemisphere.

Let's look at out data with motor imagery:

```python
motor_imagery = eegtools.io.load_edf('Team B/Olya_5sec_random_fingers_5real_and_5idle_and_5imaginary.edf')
plot_channels(motor_imagery, RHYTHMS)
```

Let's look at the channels FC5 and FC6 as they are close to premotor cortex. Let's compare their three power spectra, splitting the whole time series to three segments.

```python
plot_channels(motor_imagery, ['FC5', 'FC6'])
```

```python
def plot_single_third(data, label, sample_rate):
    f, Pxx_den = signal.welch(data, sample_rate)
    plt.plot(f[8:], Pxx_den[8:], label=label)

def plot_power_by_thirds(data, channel):
    x = data.X[data.chan_lab.index(channel)]
    N, = x.shape
    real, idle, imagery = x[:N / 3], x[N / 3:2 * N / 3], x[2 * N / 3:]
    plt.figure(figsize=(20, 12))
    plot_single_third(real, 'real', data.sample_rate)
    plot_single_third(idle, 'idle', data.sample_rate)
    plot_single_third(imagery, 'imagery', data.sample_rate)
    plt.xlabel('frequency [Hz]')
    plt.ylabel('PSD [V**2/Hz]')
    plt.legend()
    plt.plot()

plot_power_by_thirds(motor_imagery, 'FC5')
```

```python
plot_power_by_thirds(motor_imagery, 'FC6')
```
